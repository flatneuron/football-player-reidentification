```python
import cv2

# Define colors for each class
CLASS_COLORS = {
    "player": (0, 0, 0),      # black
    "referee": (255, 0, 0),     # Blue
    "ball": (0, 0, 255)         # Red
}

def draw_detections(
    frame,
    detections,
    class_names,
    show_tracking=False
):
    """
    Args:
        frame (np.ndarray): The image frame (BGR)
        detections (list of dict): Each dict contains:
            - 'box': [x1, y1, x2, y2]
            - 'conf': float
            - 'cls': int (index in class_names)
            - 'id': int (optional, if show_tracking is True)
        class_names (dict): {0: 'ball', 1: 'referee', 2: 'player'}
        show_tracking (bool): Whether to draw track IDs
    """
    for det in detections:
        box = list(map(int, det['box']))
        conf = det['conf']
        cls_id = det['cls']
        track_id = det.get('id', None)

        label = f"{class_names[cls_id]} {conf:.2f}"
        if show_tracking and track_id is not None:
            label += f" ID:{track_id}"

        color = CLASS_COLORS.get(class_names[cls_id], (255, 255, 255))

        # Draw bbox
        cv2.rectangle(frame, (box[0], box[1]), (box[2], box[3]), color, 2)

        # Measure text size
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.5
        thickness = 1
        (text_w, text_h), baseline = cv2.getTextSize(label, font, font_scale, thickness)

        # Text box outline (no fill)
        top_left = (box[0], box[1] - text_h - 6)
        bottom_right = (box[0] + text_w + 6, box[1])
        cv2.rectangle(frame, top_left, bottom_right, color, 1)

        # Put label text
        cv2.putText(frame, label, (box[0] + 3, box[1] - 3), font, font_scale, color, thickness)

    return frame
```

```python
import os
import cv2
import torch
import numpy as np
from ultralytics import YOLO
from supervision import ByteTrack, Detections
from torchreid.reid.utils.feature_extractor import FeatureExtractor
from sklearn.metrics.pairwise import cosine_similarity
from PIL import Image

# === CONFIG ===
VIDEO_PATH       = "drive/MyDrive/stealth mode/soccer_match.mp4"
MODEL_PATH       = "drive/MyDrive/stealth mode/model.pt"
REID_MODEL_PATH  = "drive/MyDrive/stealth mode/og/osnet/model/model.pth.tar-1"
OUTPUT_VIDEO_PATH= "drive/MyDrive/stealth mode/the_finalist_final_reid_output_with_tracks.mp4"
CONF_THRESH      = 0.4
SIM_THRESHOLD    = 0.6  # cosine similarity threshold

# === COLORS & DRAWING ===
CLASS_COLORS = {
    "player":  (0, 0, 0),
    "referee": (255, 0, 0),
    "ball":    (0, 0, 255)
}
def draw_detections(frame, detections, class_names, show_tracking=True):
    for det in detections:
        x1,y1,x2,y2 = map(int, det['box'])
        conf       = det['conf']
        cls_id     = det['cls']
        rid        = det.get('id')
        label      = f"{class_names[cls_id]} {conf:.2f}"
        if show_tracking and rid is not None:
            label += f" ID:{rid}"
        color = CLASS_COLORS.get(class_names[cls_id], (255,255,255))
        cv2.rectangle(frame, (x1,y1),(x2,y2), color, 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        (tw,th),_ = cv2.getTextSize(label, font, 0.5, 1)
        cv2.rectangle(frame, (x1,y1-th-6),(x1+tw+6,y1), color, 1)
        cv2.putText(frame, label, (x1+3,y1-3), font, 0.5, color, 1)
    return frame

# === INIT REID EXTRACTOR & DB ===
extractor = FeatureExtractor(
    model_name = 'osnet_x1_0',
    model_path = REID_MODEL_PATH,
    device     = 'cuda' if torch.cuda.is_available() else 'cpu'
)
player_features_db = {}  # track_id -> feature vector

# === OPEN VIDEO & SETUP OUTPUT ===
cap    = cv2.VideoCapture(VIDEO_PATH)
fps    = int(cap.get(cv2.CAP_PROP_FPS))
w      = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h      = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out    = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (w, h))

# === LOAD DETECTION & TRACKER ===
yolo         = YOLO(MODEL_PATH)
byte_tracker = ByteTrack()
class_names  = yolo.names

# === PROCESS FRAMES ===
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 1) Detect
    result = yolo(frame)[0]
    boxes  = result.boxes

    detections_to_draw = []
    if boxes is not None and len(boxes) > 0:
        det = Detections(
            xyxy      = boxes.xyxy.cpu().numpy(),
            confidence= boxes.conf.cpu().numpy(),
            class_id  = boxes.cls.cpu().numpy().astype(int)
        )
        det = det[det.confidence > CONF_THRESH]

        # 2) Track
        tracks = byte_tracker.update_with_detections(det)

        # Keep track of which ReID IDs are used this frame
        used_reid_ids = set()

        # 3) For each track, extract ReID feature & assign ID
        for i in range(len(tracks)):
            track_id = int(tracks.tracker_id[i])
            class_id = int(tracks.class_id[i])
            x1,y1,x2,y2 = map(int, tracks.xyxy[i])
            conf = float(tracks.confidence[i])

            reid_id = track_id  # default initialization

            if class_names[class_id] == 'player':
                crop = frame[y1:y2, x1:x2]
                if crop.size > 0:
                    # preprocess and extract feature
                    crop = cv2.resize(crop, (128, 256))
                    cv2.imwrite('temp.jpg', crop)
                    feature = extractor('temp.jpg')[0]


                    if player_features_db:
                        # compute similarities
                        feats = np.stack([f.cpu() for f in player_features_db.values()], axis=0)
                        ids   = list(player_features_db.keys())
                        sims  = cosine_similarity([feature.cpu()], feats)[0]
                        # sort by descending similarity
                        sorted_idxs = np.argsort(sims)[::-1]
                        # find first match not used this frame and above threshold
                        for idx in sorted_idxs:
                            if sims[idx] < SIM_THRESHOLD:
                                break
                            candidate_id = ids[idx]
                            if candidate_id not in used_reid_ids:
                                reid_id = candidate_id
                                break
                        else:
                            # no existing id free or high enough → use track_id
                            reid_id = track_id
                            player_features_db[track_id] = feature
                    else:
                        # first entry ever
                        reid_id = track_id
                        player_features_db[track_id] = feature

            # mark this id used in current frame
            used_reid_ids.add(reid_id)

            detections_to_draw.append({
                'box':  [x1, y1, x2, y2],
                'conf': conf,
                'cls':  class_id,
                'id':   reid_id
            })

    # 4) Draw and write
    frame = draw_detections(frame, detections_to_draw, class_names, show_tracking=True)
    out.write(frame)

# === CLEANUP ===
cap.release()
out.release()
print("✅ Done: saved output to", OUTPUT_VIDEO_PATH)
```

```python
import os
import cv2
import torch
import numpy as np
from ultralytics import YOLO
from supervision import ByteTrack, Detections
from torchreid.reid.utils.feature_extractor import FeatureExtractor
from sklearn.metrics.pairwise import cosine_similarity

# === 1) DEVICE SETUP ===
DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)

# === 2) CONFIG ===
VIDEO_PATH        = "drive/MyDrive/stealth mode/soccer_match.mp4"
MODEL_PATH        = "drive/MyDrive/stealth mode/model.pt"
REID_MODEL_PATH   = "drive/MyDrive/stealth mode/og/osnet/model/model.pth.tar-1"
OUTPUT_VIDEO_PATH = "drive/MyDrive/stealth mode/final_reid_output_with_tracks_gpu.mp4"
CONF_THRESH       = 0.4
SIM_THRESHOLD     = 0.6  # cosine-sim threshold

# === 3) DRAWING UTILITIES ===
CLASS_COLORS = {
    "player":  (0, 0, 0),
    "referee": (255, 0, 0),
    "ball":    (0, 0, 255)
}
def draw_detections(frame, detections, class_names, show_tracking=True):
    for det in detections:
        x1, y1, x2, y2 = map(int, det['box'])
        conf          = det['conf']
        cls_id        = det['cls']
        rid           = det.get('id')
        label         = f"{class_names[cls_id]} {conf:.2f}"
        if show_tracking and rid is not None:
            label += f" ID:{rid}"
        color = CLASS_COLORS.get(class_names[cls_id], (255,255,255))
        cv2.rectangle(frame, (x1,y1), (x2,y2), color, 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        (tw,th), _ = cv2.getTextSize(label, font, 0.5, 1)
        cv2.rectangle(frame, (x1, y1-th-6), (x1+tw+6, y1), color, 1)
        cv2.putText(frame, label, (x1+3, y1-3), font, 0.5, color, 1)
    return frame

# === 4) INIT MODELS ===
# YOLO (load on CPU; inference will use DEVICE)
yolo = YOLO(MODEL_PATH)

# OSNet Feature Extractor on DEVICE
extractor = FeatureExtractor(
    model_name='osnet_x1_0',
    model_path=REID_MODEL_PATH,
    device=DEVICE
)

# ByteTrack and feature DB
byte_tracker = ByteTrack()
player_features_db = {}  # track_id -> 512‑d numpy feature

# === 5) OPEN VIDEO & SETUP OUTPUT ===
cap  = cv2.VideoCapture(VIDEO_PATH)
fps  = int(cap.get(cv2.CAP_PROP_FPS))
w    = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h    = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out    = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (w,h))

class_names = yolo.names
frame_count = 0

# === 6) PROCESS EACH FRAME ===
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 6.1) DETECTION on GPU via callback
    result = yolo(frame, device=DEVICE)[0]
    boxes  = result.boxes

    if boxes is not None and len(boxes) > 0:
        dets = Detections(
            xyxy       = boxes.xyxy.cpu().numpy(),
            confidence = boxes.conf.cpu().numpy(),
            class_id   = boxes.cls.cpu().numpy().astype(int)
        )
        dets = dets[dets.confidence > CONF_THRESH]
    else:
        dets = Detections.empty()


    # 6.2) TRACKING
    tracks = byte_tracker.update_with_detections(dets)

    # 6.3) RE‑ID per frame
    used_ids = set()
    draw_list = []

    for i in range(len(tracks)):
        track_id = int(tracks.tracker_id[i])
        cls_id   = int(tracks.class_id[i])
        x1,y1,x2,y2 = map(int, tracks.xyxy[i])
        conf     = float(tracks.confidence[i])

        reid_id = track_id

        if class_names[cls_id] == 'player':
            crop = frame[y1:y2, x1:x2]
            if crop.size > 0:
                # preprocess and save temp
                crop_resized = cv2.resize(crop, (128,256))
                cv2.imwrite('temp.jpg', crop_resized)

                # extract feature on DEVICE
                feat = extractor('temp.jpg')[0]
                # ensure numpy on CPU
                if torch.is_tensor(feat):
                    feat = feat.cpu().numpy()

                # match against DB
                if player_features_db:
                    feats = np.stack(list(player_features_db.values()), axis=0)
                    ids   = list(player_features_db.keys())
                    sims  = cosine_similarity([feat], feats)[0]

                    for idx in np.argsort(sims)[::-1]:
                        if sims[idx] < SIM_THRESHOLD:
                            break
                        cand = ids[idx]
                        if cand not in used_ids:
                            reid_id = cand
                            break
                    else:
                        # no match → store under this track_id
                        player_features_db[track_id] = feat
                else:
                    # first ever
                    player_features_db[track_id] = feat

        used_ids.add(reid_id)
        draw_list.append({
            'box':  [x1,y1,x2,y2],
            'conf': conf,
            'cls':  cls_id,
            'id':   reid_id
        })

    # 6.4) DRAW & WRITE
    frame = draw_detections(frame, draw_list, class_names, show_tracking=True)
    out.write(frame)
    frame_count += 1

# === 7) CLEANUP ===
cap.release()
out.release()
print(f"✅ Done: wrote {frame_count} frames to {OUTPUT_VIDEO_PATH}")
```
