## Load Video

Load the soccer match video.

**Reasoning**:
Use OpenCV to load the video file for processing.

```python
import cv2

video_path = 'soccer_match.mp4'  # Replace with the actual path to your uploaded video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
else:
    print(f"Video loaded successfully from {video_path}")
    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = int(cap.get(cv2.CAP_PROP_FOURCC))

    print(f"Video properties:")
    print(f"  Width: {frame_width}")
    print(f"  Height: {frame_height}")
    print(f"  FPS: {fps}")
    print(f"  FourCC: {fourcc}")
```

## Process Video

### Subtask:
Iterate through the video frames, perform inference using the YOLOv11 model to detect players, and draw bounding boxes around detected players.

**Reasoning**:
Read frames from the video, use the loaded YOLO model to predict bounding boxes, and draw the results on the frames.

```python
import cv2
from google.colab.patches import cv2_imshow

# Define the output video path
output_video_path = 'output_soccer_match.mp4'  # Replace with your desired output path

# Define the codec and create VideoWriter object
# The FourCC code needs to be compatible with the output file format and available codecs
# 'mp4v' is a common and generally compatible codec for MP4
# out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

# if not out.isOpened():
#     print(f"Error: Could not create video writer for {output_video_path}")
# else:
#     print(f"Video writer created successfully for {output_video_path}")

# Re-initialize video capture
video_path = 'soccer_match.mp4'  # Replace with the actual path to your uploaded video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
else:
    print(f"Video loaded successfully from {video_path}")
    # Get video properties (re-get in case the object was fully reset)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = int(cap.get(cv2.CAP_PROP_FOURCC))

    print(f"Video properties:")
    print(f"  Width: {frame_width}")
    print(f"  Height: {frame_height}")
    print(f"  FPS: {fps}")
    print(f"  FourCC: {fourcc}")


    # Process the first frame
    ret, frame = cap.read()

    if ret:
        # Display the input frame
        print("Input Frame:")
        cv2_imshow(frame)

        # Perform inference
        results = model(frame)

        # Display the textual output from the model
        print("\nModel Output:")
        for r in results:
            print(r)

        # Draw bounding boxes on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        print("\nAnnotated Frame:")
        cv2_imshow(annotated_frame)


        # Release everything when job is finished
        cap.release()
        # out.release()
        # print(f"Processed video saved to {output_video_path}")
    else:
        print("Could not read the first frame from the video.")

# Release the video capture object in case of error or if not processing the whole video
if 'cap' in locals() and cap.isOpened():
    cap.release()
```
