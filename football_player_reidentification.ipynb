# Training Reid Model (skip to  soccer-player-reidentification) if dont want to train and use the already trained model

```python
import torchreid
import torch
from torchreid.reid.models import build_model
from Train.reid_trainer import MyReidDataset
```

```python

# this registers our custom dataset with torchreid for model training
torchreid.reid.data.datasets.register_image_dataset('MY-reid-dataset-new', MyReidDataset)
```

```python
ROOT_DIR = ''
# Initialize the ImageDataManager from Torchreid.
# This loads and prepares the MY-reid-dataset-new dataset located at ROOT_DIR.
# It handles image resizing to 256x256, parallel data loading with 2 workers,
# and creates training, query, and gallery loaders used for ReID training/evaluation.
datamanager = torchreid.data.ImageDataManager(
    root=ROOT_DIR,
    sources='MY-reid-dataset-new',
    transforms=None,
    workers=2,
    height=256,
    width=256
)
```

```python
model = build_model(
    name='osnet_x1_0',                # any key from __model_factory
    num_classes=datamanager.num_train_pids,  # number of training IDs (ReID classes)
    loss='softmax',                   # or 'triplet', or both as 'softmax+triplet'
    pretrained=True,                  # use ImageNet pretrained weights
    use_gpu=torch.cuda.is_available()  # move model to GPU if available
)
```

```python
# Build the optimizer for training the ReID model.
# This uses Adam optimizer with a learning rate of 0.0003,
# and applies it to all trainable parameters of the model.
optimizer = torchreid.optim.build_optimizer(
    model,
    optim='adam',
    lr=0.0003
)
```

```python
scheduler = torchreid.optim.build_lr_scheduler(
    optimizer,
    lr_scheduler='single_step',
    stepsize=20
)
```

```python
# Initialize the training engine using the softmax loss function.
# This engine handles the full training and evaluation loop for image-based Re-ID.
# - datamanager: provides training, query, and gallery dataloaders
# - model: the neural network (e.g., OSNet) to be trained
# - optimizer: optimization algorithm (e.g., Adam) to update model weights
# - scheduler: adjusts learning rate during training (optional but useful)
# - label_smooth: applies label smoothing to prevent overfitting
# - use_gpu: enables GPU acceleration if available
engine = torchreid.reid.engine.image.ImageSoftmaxEngine(
    datamanager,
    model,
    optimizer=optimizer,
    scheduler=scheduler,
    label_smooth=True,
    use_gpu=torch.cuda.is_available()
)

# Move model to GPU if available
if torch.cuda.is_available():
    model = model.cuda()


# Start the training process
# - save_dir: path to save logs, model checkpoints, and evaluation results
# - max_epoch: number of training epochs
# - eval_freq: evaluate the model on validation data every N epochs
# - print_freq: how frequently (in iterations) to print training logs
# - test_only: if True, skips training and only evaluates the model
engine.run(
    save_dir='og/osnet',
    max_epoch=1,
    eval_freq=1,
    print_freq=2,
    test_only=False
)
```

```python
from Utils.utils import (
    get_device, load_models, setup_video_io, get_detections,
    process_tracks, draw_detections
)
```

```python
# this block does inference on the input video and saves the output in OUTPUT.mp4
import os
import cv2

# === CONFIG ===
VIDEO_PATH = "inputvideo.mp4"                 #this is the 15second video that was provided for the problem 2 (reidentifying players getting out of camera frame)
MODEL_PATH = "best.pt"                        #fine tuned yolo model provided for the object detection
REID_MODEL_PATH = "reid_model.pth.tar-1"      #my osnet model for learning embeddings for the players
OUTPUT_VIDEO_PATH = "OUTPUT.mp4"              #final output video of the model showing player id for each player
CONF_THRESH = 0.4
SIM_THRESHOLD = 0.6

def main():
    device = get_device()
    print("Using device:", device)
    yolo, extractor, byte_tracker, player_features_db = load_models(MODEL_PATH, REID_MODEL_PATH, device)
    cap, out, _, _, _ = setup_video_io(VIDEO_PATH, OUTPUT_VIDEO_PATH)
    class_names = yolo.names
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        detections = get_detections(frame, yolo, device, CONF_THRESH)
        tracks = byte_tracker.update_with_detections(detections)
        draw_list = process_tracks(frame, tracks, class_names, extractor, player_features_db, SIM_THRESHOLD)
        frame = draw_detections(frame, draw_list, class_names, show_tracking=True)
        out.write(frame)
        frame_count += 1

    cap.release()
    out.release()
    print(f"wrote {frame_count} frames to {OUTPUT_VIDEO_PATH}")

if __name__ == "__main__":
    main()
```
